```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
boston.keys()
boston.feature_names
```

```python
boston
```

```python
X=pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
Y=boston.target
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=42,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc
```

```python
X_train.shape
```

```python
X_train.isna().sum()
```

```python
X_train_sc=pd.DataFrame(sc.fit_transform(X_train), columns=X_train.columns)
X_test_sc=pd.DataFrame(sc.transform(X_test), columns=X_test.columns)
```

```python
X_train_sc.describe()
```

```python
X_test_sc.describe()
```

```python
from sklearn.manifold import TSNE
```

```python
?TSNE
```

```python
X_train_tsne = TSNE(n_components=2, learning_rate=250, random_state=42).fit_transform(X_train_sc)
```

```python
X_train_tsne.shape
```

```python
X_train_tsne
```

```python
%matplotlib inline
plt.scatter(X_train_tsne[:, 0], X_train_tsne[:, 1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
?KMeans
```

```python
kmeans = KMeans(n_clusters=3, max_iter=100, random_state=42)
color_train = kmeans.fit_predict(X_train_sc)
```

```python
color_train
```

```python
plt.scatter(X_train_tsne[:, 0], X_train_tsne[:, 1], c=color_train )
plt.show()
```

```python
color_train.max()
```

```python
for i in range (color_train.max()+1):
    print (y_train[color_train==i].mean())
```

```python
for i in range (color_train.max()+1):
    print (X_train.loc[color_train==i,'CRIM'].mean())
```

```python
the_test = kmeans.predict(X_test_sc)
```

```python
the_test
```

```python
for i in range (the_test.max()+1):
    print (y_test[the_test==i].mean())
```

```python
for i in range (the_test.max()+1):
    print (X_test.loc[the_test==i,'CRIM'].mean())
```
